```python
class Atm:

    def __init__(self):
        # Initialize ATM with empty PIN and zero balance
        self.__pin = ''
        self.__balance = 0
        self.__menu()

    def __menu(self):
        # Display main menu and handle user input
        while True:
            user_input = input("""
Hello, what would you like to do?
1. Enter 1 to create pin
2. Enter 2 to deposit
3. Enter 3 to withdraw
4. Enter 4 to check balance
5. Enter 5 to exit
Your choice: """)

            if user_input == '1':
                self.create_pin()
            elif user_input == '2':
                self.deposit()
            elif user_input == '3':
                self.withdraw()
            elif user_input == '4':
                self.check_balance()
            elif user_input == '5':
                print("Thank you for using the ATM. Goodbye!")
                break
            else:
                print("Invalid option. Please try again.")

    def create_pin(self):
        # Allow user to set a new PIN
        self.__pin = input('Set your PIN: ')
        print('PIN set successfully.')

    def deposit(self):
        # Deposit funds after verifying PIN
        temp = input('Enter your PIN: ')
        if temp == self.__pin:
            try:
                amount = int(input('Enter deposit amount: '))
                self.__balance += amount
                print('Deposit successful.')
            except ValueError:
                print("Invalid input. Please enter a number.")
        else:
            print('Incorrect PIN.')

    def withdraw(self):
        # Withdraw funds after verifying PIN and checking balance
        temp = input('Enter your PIN: ')
        if temp == self.__pin:
            try:
                amount = int(input('Enter withdrawal amount: '))
                if amount <= self.__balance:
                    self.__balance -= amount
                    print('Withdrawal successful.')
                else:
                    print('Insufficient funds.')
            except ValueError:
                print("Invalid input. Please enter a number.")
        else:
            print('Incorrect PIN.')

    def check_balance(self):
        # Display current balance after verifying PIN
        temp = input('Enter your PIN: ')
        if temp == self.__pin:
            print(f'Your current balance is: ₹{self.__balance}')
        else:
            print('Incorrect PIN.')
```
